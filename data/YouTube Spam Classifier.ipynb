```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df1=pd.read_csv('Youtube01-Psy.csv', usecols=['CONTENT', 'CLASS'])
```

```python
df1.head()
```

```python
df2=pd.read_csv('Youtube02-KatyPerry.csv', usecols=['CONTENT', 'CLASS'])
df3=pd.read_csv('Youtube03-LMFAO.csv', usecols=['CONTENT', 'CLASS'])
df4=pd.read_csv('Youtube04-Eminem.csv', usecols=['CONTENT', 'CLASS'])
df5=pd.read_csv('Youtube05-Shakira.csv', usecols=['CONTENT', 'CLASS'])
```

```python
# Concatinating all the dataframes into 1

df = pd.concat([df1,df2,df3,df4,df5])
```

```python
df.shape
```

### Splitting the data into target and independent

```python
X = df['CONTENT']
y = df['CLASS']
```

### Feature extraction from text data

```python
cv = CountVectorizer()
```

```python
X = cv.fit_transform(X)
```

```python
#Getting feature names
cv.get_feature_names()
```

### Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Model Building

```python
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred=nb.predict(X_test)
accuracy_score(y_test, y_pred)
```

### Pickling our model

```python
# Pickle file for Naive Bayes model

filename = 'naive_bayes.pkl'
pickle.dump(nb, open(filename, 'wb'))
```

```python
# Pickle file for the CountVectorizer object

filename = 'vectorizer.pkl'
pickle.dump(cv, open(filename, 'wb'))
```
